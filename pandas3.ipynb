```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
csv_path = "C:\\Users\\HP\\Downloads\\reports_2025-01_current_stock_v2_232707_1737160618852.csv"
df = pd.read_csv(csv_path)

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Print available columns for debugging
print("🔹 Available Columns:", df.columns.tolist())

# Identify the correct columns
company_column = "Distributor Name"  
value_column = "Total Value"  # Found in the dataset

# Ensure the "Total Value" column is numeric
df[value_column] = pd.to_numeric(df[value_column], errors='coerce')

# Drop NaN values in the selected columns
df = df.dropna(subset=[value_column])

# Debugging: Print first few rows
print("✔ First 5 Rows:\n", df[[company_column, value_column]].head())

# Group by "Distributor Name" and sum the total value
df_grouped = df.groupby(company_column)[value_column].sum()

# Plot the bar graph
if df_grouped.empty:
    print("❌ No Data Available for Plotting!")
else:
    plt.figure(figsize=(12, 6))
    df_grouped.plot(kind="bar", color="blue")
    plt.xlabel("Distributor Name")
    plt.ylabel("Total Value")
    plt.title("Distributor-wise Total Value Distribution")
    plt.xticks(rotation=45)
    plt.show()
```
